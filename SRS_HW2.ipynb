##### -   symboling (safety is called symbolling here, integer from -2 - the safest to +3 - the riskiest)

##### -   doornumber (two or four)
##### -   carbody (sedan/hatchback/wagon/hardtop)

##### -   carwidth

##### -   curbweight (the total mass of a vehicle with standard equipment and all necessary operating consumables)
##### -   enginetype (ohc/dohc/ohcv/dohcv/ohef/l/rotor)
##### -   cylindernumber (two/three/four/five/six/eight/twelve)
##### -   enginesize
##### -   fuelsystem (mpfi/m/4bbl/2bb/1bb/idi/spdi/spfi)
##### -   boreratio (the ratio between the dimensions of the engine cylinder bore diameter to its piston stroke-length)
##### -   stroke
##### -   compressionratio (the ratio of the volume of the cylinder and the combustion chamber when the piston is at the bottom, and the volume of the combustion chamber when the piston is at the top)
##### -   horsepower
##### -   peakrpm (how many times the engine's crankshaft makes one full rotation every minute, and along with it, how many times each piston goes up and down in its cylinder)
##### -   citympg (approximately how far car will travel using a single UK gallon of fuel in city)
##### -   highwaympg (mpg for highway)


```python
df2 = df1.copy()
```

```python
for coll1 in ['fueltype_num', 'aspiration_num', 'doornumber']:
    a = df2[coll1].unique()
    df2[coll1] = df2[coll1].replace({a[0]: 0, a[1]: 1})

print(df2)
```

```python
df3 = df2.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore')

for coll2 in ['carbody_num', 'drivewheel_num', 'enginetype_num', 'cylindernumber', 'fuelsystem_num']:
    enc = pd.DataFrame(encoder.fit_transform(df3[[coll2]]).toarray(), columns=encoder.categories_)
    enc.columns = [f'{coll2}_' + str(col).replace('(', '').replace(')', '').replace(',', '') for col in enc.columns]
    df3 = df3.join(enc.astype(int))
    df3.drop(coll2, axis= 1, inplace= True)

df3.drop('car_ID', axis=1, inplace=True)
df3
```

```python
from sklearn.model_selection import train_test_split

df4 = df3.copy()
y = np.array(df4['price'])
df4.drop('price', axis= 1, inplace= True)


df_train, df_test, price_train, price_test = train_test_split(df4, y, test_size=0.3, random_state=np.random.seed(50))
```

## Модель 1

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(df_train, price_train)

pred_tr = model.predict(df_train)
pred_ts = model.predict(df_test)

mse_tr = mean_squared_error(price_train, pred_tr)
mse_ts = mean_squared_error(price_test, pred_ts)

print("MSE train set:", mse_tr)
print("MSE test set:", mse_ts)
```

```python
coeffs = model.coef_
print(np.array2string(coeffs, formatter={'float_kind':lambda x: f'{x:.3f}'}))
```

```python
lstt = sorted(list(zip(df_train.columns, coeffs)),
       key=lambda x: abs(x[1]))
for el in lstt:
  print(f'{el[0]}: {el[1]:.3f}')
```

## Гетероскедастичность

```python
intercept = model2.intercept_
residuals = price_train_new3 - pred_tr3

train = df_train_new3.copy()
train['intercept'] = intercept

train
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

bp_test = sm.stats.diagnostic.het_breuschpagan(residuals, train)
print('Breusch-Pagan test:')
print('LM statistic:', bp_test[0])
print('LM p-value:', bp_test[1])

plt.scatter(pred_tr3, residuals, c ='forestgreen')
plt.axhline(0, linestyle='--', color='purple')
plt.xlabel('Предсказанные значения')
plt.ylabel('Остатки')
plt.title('График остатков')
plt.show()
```

## Случайный лес

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
max_depth = [i for i in range(1, 100)]
max_samples = [i for i in range(1, 100)]

mse = []

for d in max_depth:
    for l in max_samples:
        rfr = RandomForestRegressor(max_depth=d, max_samples=l)
        rfr.fit(df_train, price_train)

        train_pred_y = rfr.predict(df_train)
        test_pred_y = rfr.predict(df_test)

        mse.append([mean_squared_error(test_pred_y, price_test), mean_squared_error(train_pred_y, price_train), d, l])

mse1 = sorted(mse, key=lambda x: (x[0], x[1]))

print(mse1[0])
```

```python
rfr = RandomForestRegressor(max_depth=59, max_samples=77)
rfr.fit(df_train, price_train)

train_pred_y2 = rfr.predict(df_train)
test_pred_y2 = rfr.predict(df_test)

print(mean_squared_error(train_pred_y2, price_train))
print(mean_squared_error(test_pred_y2, price_test))
```

```python
import matplotlib.pyplot as plt

importances = rfr.feature_importances_
feature_importance_df = pd.DataFrame({'Feature': df_train.columns, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)


plt.figure(figsize=(10, 10))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'], color = 'forestgreen')
plt.xlabel('Importance')
plt.title('Feature Importance')
plt.show()
```
